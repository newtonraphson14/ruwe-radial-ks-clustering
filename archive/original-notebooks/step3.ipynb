```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Data
df = pd.read_csv("M67_Members_Final.csv")

# Pastikan data angka
cols = ['bp_rp', 'phot_g_mean_mag', 'ruwe']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(subset=cols)

# 2. Plot CMD dengan Warna RUWE
plt.figure(figsize=(10, 8))

# Trik Visualisasi:
# Kita urutkan data biar nilai RUWE tinggi digambar TERAKHIR (di paling atas tumpukan)
# Supaya titik merahnya gak ketiban titik biru.
df_sorted = df.sort_values('ruwe')

# Scatter Plot
# c = ruwe (warna berdasarkan RUWE)
# vmin=0.8, vmax=1.4 (Fokus kontras di area kritis: 1.0 = bagus, 1.4 = biner/jelek)
sc = plt.scatter(df_sorted['bp_rp'], df_sorted['phot_g_mean_mag'], 
                 s=3, c=df_sorted['ruwe'], cmap='turbo', vmin=0.8, vmax=1.4, alpha=0.8)

# --- ESTETIKA ---
plt.gca().invert_yaxis()
cbar = plt.colorbar(sc)
cbar.set_label('Gaia RUWE Score', fontsize=12)

plt.xlabel("Color ($G_{BP} - G_{RP}$)")
plt.ylabel("Absolute Magnitude ($G$)")
plt.title("M67 CMD Colored")
plt.grid(True, alpha=0.3)

plt.show()
```
