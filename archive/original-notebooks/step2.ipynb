```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load Data
try:
    df = pd.read_csv("M67_Members_Final.csv")
    print(f"✅ Data Loaded: {len(df)} stars")
except:
    print("❌ File gak ketemu. Pastikan step sebelumnya sukses.")

# --- PERBAIKAN: PAKSA JADI ANGKA (ANTI LAG) ---
# Ini langkah krusial yang bikin plot tadi macet
df['bp_rp'] = pd.to_numeric(df['bp_rp'], errors='coerce')
df['phot_g_mean_mag'] = pd.to_numeric(df['phot_g_mean_mag'], errors='coerce')

# Buang data yang jadi NaN setelah dipaksa
cmd_data = df.dropna(subset=['bp_rp', 'phot_g_mean_mag'])
print(f"✅ Data Siap Plot: {len(cmd_data)} stars (Tipe data sudah benar)")

# 2. Plotting CMD
plt.figure(figsize=(8, 8))

# Scatter Plot
plt.scatter(cmd_data['bp_rp'], cmd_data['phot_g_mean_mag'], 
            s=2, c='black', alpha=0.6, label='Members')

# --- ESTETIKA ---
plt.gca().invert_yaxis() # Wajib dibalik buat Magnitudo
plt.xlabel("Color ($G_{BP} - G_{RP}$)", fontsize=12)
plt.ylabel("Absolute Magnitude ($G$)", fontsize=12)
plt.title("M67 Color-Magnitude Diagram (Final Members)", fontsize=14)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
